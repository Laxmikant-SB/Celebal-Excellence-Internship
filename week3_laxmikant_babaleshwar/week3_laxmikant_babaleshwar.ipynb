## 6) Basic cleaning

Handle missing values, remove duplicates, and make sure numeric columns are clean.

```python
df = df.copy()
df.columns = [c.strip().lower() for c in df.columns]
df = df.drop_duplicates()

for col in df.columns:
    if col != 'country':
        df[col] = pd.to_numeric(df[col], errors='coerce')

numeric_cols = [c for c in df.columns if c != 'country']
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].median())

df.isna().sum()
```

## 7) Exploratory Data Analysis

```python
plt.figure(figsize=(12, 8))
sns.heatmap(df.select_dtypes(include=np.number).corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

```python
for col in numeric_cols:
    plt.figure(figsize=(8, 3))
    sns.boxplot(x=df[col])
    plt.title(f'Boxplot: {col}')
    plt.show()
```

## 8) Feature scaling

Clustering works better when features are on the same scale.

```python
features = df.drop(columns=['country'], errors='ignore')
scaler = StandardScaler()
X_scaled = scaler.fit_transform(features)
X_scaled.shape
```

## 9) K-Means: Elbow method

```python
inertias = []
k_values = range(2, 11)

for k in k_values:
    model = KMeans(n_clusters=k, random_state=42, n_init=10)
    model.fit(X_scaled)
    inertias.append(model.inertia_)

plt.figure(figsize=(8, 4))
plt.plot(list(k_values), inertias, marker='o')
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()
```

## 10) Train K-Means

```python
best_k = 3  # change after checking the elbow plot
kmeans = KMeans(n_clusters=best_k, random_state=42, n_init=10)
df['kmeans_cluster'] = kmeans.fit_predict(X_scaled)

print('Silhouette Score:', silhouette_score(X_scaled, df['kmeans_cluster']))
df[['country', 'kmeans_cluster']].head() if 'country' in df.columns else df[['kmeans_cluster']].head()
```

## 11) Try DBSCAN

```python
dbscan = DBSCAN(eps=1.5, min_samples=5)
df['dbscan_cluster'] = dbscan.fit_predict(X_scaled)

print(df['dbscan_cluster'].value_counts().sort_index())
```

## 12) PCA visualization

```python
pca = PCA(n_components=2, random_state=42)
X_pca = pca.fit_transform(X_scaled)

viz = pd.DataFrame({
    'pca1': X_pca[:, 0],
    'pca2': X_pca[:, 1],
    'cluster': df['kmeans_cluster']
})

plt.figure(figsize=(8, 6))
sns.scatterplot(data=viz, x='pca1', y='pca2', hue='cluster', palette='tab10')
plt.title('K-Means Clusters Visualized with PCA')
plt.show()
```

## 13) Cluster profiling

```python
profile = df.groupby('kmeans_cluster')[numeric_cols].mean().round(2)
profile
```

## 14) Final insights

Based on the K-Means clustering (k=3, silhouette score ≈ 0.28) and the cluster profiling table above:

- **Cluster 1 — high child mortality / low development**: Average `child_mort` ≈ 93 deaths per 1,000 live births (far higher than the other clusters), `total_fer` ≈ 5.0 children per woman, `life_expec` ≈ 59 years (lowest), `income` ≈ $3,942 and `gdpp` ≈ $1,922 (both lowest), and the highest `inflation` (≈12%). This cluster represents poor, underdeveloped countries with weak health outcomes.
- **Cluster 0 — high income / developed economies**: `income` ≈ $45,672 and `gdpp` ≈ $42,494 (both by far the highest), `child_mort` ≈ 5 (lowest), and `life_expec` ≈ 80 years (highest). This cluster represents wealthy, developed nations.
- **Cluster 2 — middle-income / emerging economies**: Sits between the other two on almost every metric — `income` ≈ $12,306, `gdpp` ≈ $6,486, `child_mort` ≈ 22, `life_expec` ≈ 73 years. These are emerging economies that are neither fully developed nor in acute crisis.
- **Underdeveloped cluster**: Cluster 1 clearly looks the most underdeveloped — it combines the lowest income/gdpp with the worst health and demographic indicators (highest child mortality and fertility, lowest life expectancy) and the highest inflation, suggesting economic instability on top of weak health infrastructure.
- **Countries to prioritize for aid**: Countries in **Cluster 1** should be prioritized for developmental and health aid, since they show the weakest economic indicators (low `gdpp`/`income`) combined with the worst health outcomes (high `child_mort`, high `total_fer`, low `life_expec`) of all three groups.
